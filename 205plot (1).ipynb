```python
%matplotlib inline
from urllib import urlopen

import brewer2mpl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Set up some better defaults for matplotlib
from matplotlib import rcParams

#colorbrewer2 Dark2 qualitative color table
dark2_colors = brewer2mpl.get_map('Dark2', 'Qualitative', 7).mpl_colors

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'


def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()

```

```python
def plot_speedup(sizes, serial_times, base, grandchild, depth, dff, num_proc):
#     plt.figure(figsize(10,15))
#     plt.subplot(211)
    plt.plot(sizes, [s/c for c,s in zip(base[num_proc], serial_times)], label='Base')
    plt.plot(sizes, [s/c for c,s in zip(grandchild[num_proc], serial_times)], label='Grandchild')
    plt.plot(sizes, [s/c for c,s in zip(depth[num_proc], serial_times)], label='Depth First')
    plt.plot(sizes, [s/c for c,s in zip(dff[num_proc], serial_times)], label='Depth First First')
    plt.xlabel('Problem Size')
    plt.xlim(sizes[0], sizes[-1])
    plt.xticks(sizes)
    plt.ylabel('Speedup')
    plt.legend(loc='upper left')
    plt.title('Speedups of Parallel Implementation on %d Processes' % num_proc)
    remove_border()
    #plt.show()
    #savefig
     
    
```

```python
plot_speedup(sizes, serial_times, base, grandchild, depth, dff, 64)
```

```python
def plot_efficiency_for_problem_size(sizes, proc_sizes, size_idx, serial_times, base, grandchild, depth, dff):
    plt.plot (proc_sizes, [(serial_times[size_idx]/(base[np][size_idx]*np)) for np in proc_sizes], label='Base')
    plt.plot (proc_sizes, [(serial_times[size_idx]/(grandchild[np][size_idx]*np)) for np in proc_sizes], label='Grandchild')
    plt.plot (proc_sizes, [(serial_times[size_idx]/(depth[np][size_idx]*np)) for np in proc_sizes], label='Depth')
    plt.plot (proc_sizes, [(serial_times[size_idx]/(dff[np][size_idx]*np)) for np in proc_sizes], label='Depth First First')

    plt.xlabel('Number of Processes')
    plt.xlim(proc_sizes[0], proc_sizes[-1])
    plt.xticks(proc_sizes)
    plt.ylabel('Efficiency')
    plt.legend(loc='upper right')
    plt.title('Efficiency of Parallel Implementation with Problem of Size %d' % sizes[size_idx])
    remove_border()
#     #plt.show()    
```

```python
plot_efficiency_for_problem_size(sizes, proc_sizes, 5, serial_times, base, grandchild, depth, dff)
```

```python
def plot_efficiency_for_impl(sizes, proc_sizes, serial_times, impl_name, impl_data):
    
    for i,size in enumerate(sizes):
        plt.plot(proc_sizes, [(serial_times[i]/(impl_data[np][i]*np)) for np in proc_sizes], label=str(size))
                 
    plt.xlabel('Number of Processes')
    plt.xlim(proc_sizes[0], proc_sizes[-1])
    plt.xticks(proc_sizes)
    plt.ylabel('Efficiency')
    plt.legend(loc='upper right')
    plt.title('Efficiency of %s Parallel Implementation across Problem Sizes' % impl_name)
    remove_border()
```

```python
plot_efficiency_for_impl(sizes, proc_sizes, serial_times, 'Grandchildren', grandchild)
```
